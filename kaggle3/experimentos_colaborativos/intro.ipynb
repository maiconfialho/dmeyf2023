```python
def hot_deck_impute(df, missing_column, similarity_columns):
    for index, row in df.iterrows():
        if pd.isna(row[missing_column]):
            similarity_scores = []
            for index2, row2 in df.iterrows():
                if index != index2 and not pd.isna(row2[missing_column]):
                    similarity = 0
                    for column in similarity_columns:
                        if isinstance(row[column], (int, float)) and isinstance(row2[column], (int, float)):
                            similarity += abs(row[column] - row2[column])
                    similarity_scores.append((index2, similarity))

            if similarity_scores:
                similarity_scores.sort(key=lambda x: x[1])
                most_similar_index = similarity_scores[0][0]
                df.at[index, missing_column] = df.at[most_similar_index, missing_column]

missing_column = 'cat1'
similarity_columns = ['cat2']

hot_deck_impute(df, missing_column, similarity_columns)
```
